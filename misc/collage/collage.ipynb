```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
orig = Image.open("orig.png")
truth = Image.open("truth.png")
vanilla_unet = Image.open("vanilla_unet.png")
with_dd = Image.open("with_dd.png")
with_edge_mask = Image.open("with_edge_mask.png")
with_post_processing = Image.open("with_post_processing.png")
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(1, 6, 1)
plt.imshow(orig)
plt.title("Original Image")

plt.subplot(1, 6, 2)
plt.imshow(truth)
plt.title("Ground Truth")

plt.subplot(1, 6, 3)
plt.imshow(vanilla_unet)
plt.title("Vanilla U-Net")

plt.subplot(1, 6, 4)
plt.imshow(with_dd)
plt.title("With DenseNet")

plt.subplot(1, 6, 5)
plt.imshow(with_edge_mask)
plt.title("With Edge Mask")

plt.subplot(1, 6, 6)
plt.imshow(with_post_processing)
plt.title("With Post Processing")
```
